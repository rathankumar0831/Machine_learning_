```python
# minimum, Q1, Median, Q3, IQR
```

```python
lst_marks = [22, 32, 43, 24, 56, 65, 67, 87, 81, 99, 97, 85, 21]
```

```python
import numpy as np
minimum, Q1, median, Q3, maximum = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])
```

```python
minimum, Q1, median, Q3, maximum
```

```python
# Interquartile range
IQR = Q3 - Q1
IQR
```

```python
lower_fence = Q1-1.5*(IQR)
```

```python
higher_fence = Q3+1.5*(IQR)
```

```python
lower_fence
```

```python
higher_fence
```
